```python
import sqlite3
import pandas as pd
import numpy as np
import os
import sys
from datetime import timedelta, date, datetime
```

```python
from pandas.api.types import is_string_dtype
from pandas.api.types import is_numeric_dtype
```

```python
db = sqlite3.connect("../agg-data/unit-profile.db")
```

```python
df = pd.DataFrame(columns=['unit_id', 'isp', 'download', 'upload', 'technology', 'state', 'region', 'tz_offset', 'tz_offset_dst', 'year', 'month'])
for unit_profile in [i for i in os.listdir() if i.startswith('Unit-Profile-')]:
    print(unit_profile)
    df_tmp = pd.read_excel(unit_profile)
    if unit_profile == 'Unit-Profile-Sept12.xls':
        df_tmp.rename(columns={'Download':'download', 'Upload':'upload', 'Technology':'technology', 'State':'state', 'Region':'region','UTC_OFFSET':'tz_offset','UTC_OFFSET_DST':'tz_offset_dst'}, inplace=True)
        df_tmp['year'] = '2012'
        df_tmp['month'] = '09'
        
    if unit_profile == 'Unit-Profile-Sept13.xlsx':
        df_tmp.rename(columns={'DOWN':'download', 'UP':'upload', 'TECHNOLOGY':'technology', 'STATE':'state', 'REGION':'region','timezone_offset':'tz_offset','timezone_offset_dst':'tz_offset_dst'}, inplace=True)
        df_tmp['year'] = '2013'
        df_tmp['month'] = '09'
        
    if unit_profile == 'Unit-Profile-Sept14.xls':
        df_tmp.rename(columns={'SK down':'download', 'SK UP':'upload', 'TECHNOLOGY':'technology', 'STATE':'state', 'REGION':'region','timezone_offset':'tz_offset','timezone_offset_dst':'tz_offset_dst'}, inplace=True)
        df_tmp['year'] = '2014'
        df_tmp['month'] = '09'
        
    if unit_profile == 'Unit-Profile-Apr12.xlsx':
        df_tmp.rename(columns={'UNIT_ID':'unit_id', 'ISP':'isp', 'DOWNLOAD_TIER':'download', 'UPLOAD_TIER':'upload', 'TECHNOLOGY':'technology', 'STATE':'state', 'REGION':'region','UTC_OFFSET':'tz_offset','UTC_OFFSET_DST':'tz_offset_dst'}, inplace=True)
        df_tmp['year'] = '2012'
        df_tmp['month'] = '04'
        
    if unit_profile == 'Unit-Profile-Sept15.xlsx':
        df_tmp.rename(columns={'Download':'download', 'Upload':'upload', 'TECHNOLOGY':'technology', 'STATE':'state', 'CENSUS':'region','timezone_offset':'tz_offset','timezone_offset_dst':'tz_offset_dst'}, inplace=True)
        df_tmp['year'] = '2015'
        df_tmp['month'] = '09'
        
    if unit_profile == 'Unit-Profile-Sept16.xlsx':
        df_tmp.rename(columns={'Download':'download', 'Upload':'upload', 'Technology':'technology', 'State':'state', 'Census':'region','timezone_offset':'tz_offset','timezone_offset_dst':'tz_offset_dst'}, inplace=True)
        df_tmp['year'] = '2016'
        df_tmp['month'] = '09'
        
    if unit_profile == 'Unit-Profile-Sept17.xlsx':
        df_tmp.rename(columns={'ISP':'isp', 'Download':'download', 'Upload':'upload', 'Technology':'technology', 'State':'state', 'Census':'region','timezone_offset':'tz_offset','timezone_offset_dst':'tz_offset_dst'}, inplace=True)
        df_tmp['year'] = '2017'
        df_tmp['month'] = '09'
        df_tmp['tz_offset'] = df_tmp['tz_offset'].str.replace(r' hr$', '')
        df_tmp['tz_offset_dst'] = df_tmp['tz_offset_dst'].str.replace(r' hr$', '')
        df_tmp['tz_offset'] = pd.to_numeric(df_tmp['tz_offset'])
        df_tmp['tz_offset_dst'] = pd.to_numeric(df_tmp['tz_offset_dst'])
    
    
    print(df_tmp.columns.tolist())
    print(df.columns.tolist())
    df_tmp['technology'] = df_tmp['technology'].str.upper()
    df_tmp['state'] = df_tmp['state'].str.upper()
    df_tmp['region'] = df_tmp['region'].str.upper()
    df_tmp = df_tmp[['unit_id', 'isp', 'download', 'upload', 'technology', 'state', 'region', 'tz_offset', 'tz_offset_dst', 'year', 'month']]
    df = pd.concat([df, df_tmp], join='inner', axis=0, ignore_index=True)
```
